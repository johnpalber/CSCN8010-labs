```python
import pandas as pd
```

```python
housing_df = pd.read_csv("data\housing.csv", sep= ',')
housing_df.describe
```

```python
import seaborn as sns

sns.scatterplot(housing_df, x="median_income", y="median_house_value")
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

Train the model



```python
model.fit(X=housing_df[["median_income"]], y=housing_df[["median_house_value"]])
```

```python
# print coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
model.predict(X= [[12], [9], [1]])
```

```python
predicted = 12*model.coef_+model.intercept_
print(predicted)
```

```python
import matplotlib.pyplot as plt
```

```python
# plot the regression line against the training data
sns.scatterplot(x="median_income", y="median_house_value", data=housing_df, color="blue")
plt.plot(housing_df[["median_income"]], model.predict(housing_df[["median_income"]]), color="black", linewidth=3)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print(f"Mean squared error: {mean_squared_error(housing_df[['median_house_value']], model.predict(housing_df[['median_income']]))}")
print(f"Mean absolute error: {mean_absolute_error(housing_df[['median_house_value']], model.predict(housing_df[['median_income']]))}")
```

```python
sns.scatterplot(x="median_income", y="median_house_value", data=housing_df, color="red")
plt.plot(housing_df[["median_income"]], model.predict(housing_df[["median_income"]]), color="black", linewidth=3)
```
